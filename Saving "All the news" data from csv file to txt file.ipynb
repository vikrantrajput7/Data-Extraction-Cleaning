```python
import pandas as pd
```

```python
df=pd.read_csv("Headline Data.csv")
```

```python
df.head()
```

```python
df
```

```python
import re
count=0
art=[]
head=[]
for i in range(len(df["Headline"])):
    if len(df["Headline"][i]) >= 15:
        x=df["Article"][i]
        y=df["Headline"][i]
        art.append(re.sub(r'\s',' ',x))
        head.append(re.sub(r'\s',' ',y))
        count+=1
print(count)
```

```python
df=pd.DataFrame(art,columns=["Article"])
df["Headline"]=head
df.head()
```

```python
df.to_csv("Headline Data.csv",index=None)
```

```python
ndf=df.drop("Headline",axis=1)
ndf.to_csv("src.txt",index=None,header=None)
```

```python
ndf=df.drop("Article",axis=1)
ndf.to_csv("tgt.txt",index=None,header=None)
```
