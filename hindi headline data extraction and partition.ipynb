```python
src=open("news.txt",'r')
tgt=open("headline.txt",'r')
src_train=open("src-train.txt",'w')
src_val=open("src-val.txt",'w')
src_test=open("src-test.txt",'w')
tgt_train=open("tgt-train.txt",'w')
tgt_val=open("tgt-val.txt",'w')
tgt_test=open("tgt-test.txt",'w')
count=1
for line in src:
    if count <= 583903:
        src_train.write(line)
        count+=1
    else:
        if count <=588903:
            src_val.write(line)
            count+=1
        else:
            src_test.write(line)
            count+=1
src.close()
src_train.close()
src_val.close()
src_test.close()

count=1
for line in tgt:
    if count <= 583903:
        tgt_train.write(line)
        count+=1
    else:
        if count <=588903:
            tgt_val.write(line)
            count+=1
        else:
            tgt_test.write(line)
            count+=1
tgt.close()
tgt_train.close()
tgt_val.close()
tgt_test.close()
```
