```python
import os 
dirct=[]
for i in os.listdir():
    if "D" in i:
        dirct.append(i)
```

```python
import os
import re
from bs4 import BeautifulSoup
cmplt=open("all-test-data.txt",'w')
for folder in dirct:
    docs=os.listdir(folder)
    for doc in docs:
        print(doc)
        sent=[]
        with open(folder+'//'+doc,'r') as duc:
            soup=BeautifulSoup(duc,'lxml')
            if soup.headline != None:
                for data in soup.find_all('text'):
                    sent.append(re.sub('\n','',(data.text)))
                cmplt.write((re.sub('\n','',doc))+'\t')
                cmplt.write((re.sub('\n','',soup.headline.text))+'\t')
                cmplt.write(" ".join(sent)+'\n')
cmplt.close()   
```
